DESCRIPCIÓN:
La prueba de rachas se utiliza para determinar si una secuencia de datos es aleatoria. En este código se cuenta una racha cada vez que hay un cambio de signo, dependiendo de si es mayor o menor del número anterior, de este modo si el número siguiente es mayor se le asigna +, si es menor entonces -. De tal manera que cuando hay un cambio de signo se dice que existe una racha.

Nuestras hipótesis son:

$H_0$= Los números son aleatorios

$H_1$= Los núimeros no son aleatorios

Para nuestra prueba en el punto crítico usamos stats.norm.ppf(1-0.05/2), de este modo si nuestro est. de prueba se sale sabemos que se rechaza $H_0$.

Nosotros sacamos el valor absoluto del estadístico, porque es una prueba de dos colas, esto lo hace simétrico, por lo que si el estadístico es negativo y se sale se rechaza $H_o$, y si este es el mismo número pero positivo, también rebazaría nuestro punto crítico positivo.

**Nota:** Desde un inicio nosotros lo pusimos (1-0.05/2) para que fuese positivo, ya de ahí sólo tomamos z_o con su valor absoluto.

```python
def rachitas(D):  #función de prueba de rachas
    n = len(D)
    Cerox = [] #aquí almacenamos las comparaciones
    conteo = 1 #comienza en uno porque la primera ya la contamos
    for i in range(0, len(D)-1):   # Si el siguiente número es mayor se le asigna 1, si es menor ent cero
        if D[i+1] > D[i]:
            Cerox.append(1)
        else:
            Cerox.append(0)

    for i in range(len(Cerox)-1):  # cada vez que hay un cambio de signo (o en este caso de 0 a 1, o de 1 a 0) incrementa el contador de rachas
        if Cerox[i] != Cerox[i+1]:
            conteo += 1

    z_n = stats.norm.ppf(1-0.05/2) #calculamos el punto crítico
    Media = (2*n + 1)/3
    Varianza = (16*n - 29)/90
    z_0 = (conteo - Media)/np.sqrt(Varianza)  #estadístico de prueba

    if abs(z_0) > (z_n): #comparación del estadístico con el punto crítico
        print("Se rechaza H0")
    else:
        print("No se rechaza H0")

    return conteo
```
