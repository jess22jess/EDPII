Calculamos el estadístico de prueba \( D \), a partir de:

$ D^+ = \max_{1 \leq i \leq n} \left| \frac{i}{n} - x_i\right| , \quad D^- = \max_{1 \leq i \leq n} \left| x_i - \frac{i-1}{n} \right|  $

$ D = \max(D^+, D^-) $

Ya que calculamos el estadístico, entonces comparamos con el valor crítico

$ \text{Si } D < d_{(\alpha, n)} \quad \text{no se rechaza } H_0 $

```python
def pruebak(D):
  D.sort() # ordenamos los datos de la lista D de menor a mayor
  n =len(D)
  Dmas = []
  Dmenos = []
  for i in range(1,n):
   Dmas.append(abs(i/n - D[i])) # aquí calculamos los estadísticos
   Dmenos.append(abs(D[i] - (i-1)/n));
  A = Dmas + Dmenos  # Concatena las dos listas, y así tener todas las diferencias calculadas
  Dmax = max(A)
  KstestResult, D_0 = stats.kstest(D, 'uniform') # para comparar con la distribución uniforme lo que tenemos de datos
  alpha = 0.05
  if D_0 < Dmax:
       a= 'No existen suficientes pruebas para rechazar H0'
  else:
      a= 'Existen suficientes evidencias para rechazar H0'
  return a
```
